```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
```

```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.110 Safari/537.36"

options = Options()
#The --no-sandbox argument is passed to the web browser to disable the sandboxing feature.
options.add_argument('--no-sandbox')
#The --headless argument is used to run the web browser in headless mode which means that the browser operates without a visible user interface, to make it suitable for automated tasks. 
options.add_argument('--headless')
options.add_argument(f"user-agent={user_agent}") 
#The --disable-dev-shm-usage argument is used to disable the use of a /dev/shm shared memory file system in the browser.
options.add_argument('--disable-dev-shm-usage')

service = Service(r'C:\Users\awake\Downloads\chromedriver_win32\chromedriver.exe')

driver = webdriver.Chrome(service = service, options = options)

```

```python
def extract_details(url):
    driver.get(url)
    time.sleep(2)
    name = " ".join(driver.find_element(By.XPATH,"/html/body/div[3]/header[1]/h1").text.split(" ")[:-2])
    model_name = f"Device_Name : {name}"
    specs = driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div[2]')
    #specs.insert(0,model_name)
    return specs
```

```python
og_details = extract_details("https://www.devicespecifications.com/en/model/b1b25cf7")
```

```python
og_details.text
```

```python
disp_url = "https://www.devicespecifications.com/en/model-display/cde35a55"
driver.get(disp_url)
#Find the number of table sections 
sections = driver.find_elements(By.XPATH,'/html/body/div[3]/div[4]/div')
num_sections = len(sections)
#Find the number of similar phones in each section
rows = driver.find_elements(By.XPATH,'/html/body/div[3]/div[4]/div[1]/table/tbody/tr')
num_rows = len(rows)
```

```python
i = 1
#Contains list of links for similar phones
src_list = []
while (i<=num_sections):
    row_list = []
    j = 1
    while (j<=num_rows):
        #Iteratively replacing xpath with i and j
        try:
            src_element = driver.find_element(By.XPATH,"/html/body/div[3]/div[4]/div["+str(i)+"]/table/tbody/tr["+str(j)+"]/td[1]/div[1]")
            src_url = src_element.get_attribute('data-src')
            model_no = src_url.split("/")[5]
            url = f'https://www.devicespecifications.com/en/model/{model_no}'
        #details = extract_details(url)
            row_list.append(url)
        except:
            pass
        j+=1
    i+=1
    src_list.extend(row_list)    
```

```python
len(src_list)
```

```python
similar_list = [extract_details(i) for i in src_list]
```

```python
details_array = {"Phone Specifications":og_details, "Similar Phone Specifications": similar_list}
```
